## REMOVING B AND M AND USD SIGNS

```python
df2021['Market Value'][0][-1]
```

```python
def convert(df,col):
    value=[]
    for i in df[col]:
        if i[-1]=='B':
            value.append(float(i.replace('B','').replace(',','').replace('$','')))
        elif i[-1]=='M':
            value.append(float(i.replace('M','').replace(',','').replace('$',''))/1000)
        else:
            value.append(0)
    df[col]=value       
```

```python
for i in df2021.columns.tolist()[2:-1]:
    convert(df2021,i)
```

```python
for i in df2020.columns.tolist()[2:-1]:
    convert(df2020,i)
```

```python
merged_data=pd.concat([df2017,df2018,df2019,df2020,df2021]).sort_values(by='Name').reset_index(drop=True)
```
